# Relatório de Análise VIII

## Identificando e Removendo Outliers

```python
data = pd.read_csv('Exportações dos Notebooks/aluguel_residencial_excluindo_NA.csv', sep= ';')
```

```python
data.boxplot(['Valor'])
```

```python
data[data['Valor'] >= 500000]
```

```python
valor = data['Valor']
```

<img src='Box-Plot.png' width=70%>

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >= limite_inferior) & (valor <= limite_superior)
data_new = data[selecao]
```

```python
data_new.boxplot(['Valor'])
```

```python
data.hist(['Valor'])
data_new.hist(['Valor'])
```

```python
#Exercicio extra
data_exer = pd.read_csv('aluguel_amostra.csv', sep =';')
m2 = data_exer['Valor m2']
m2_Q1 = m2.quantile(.25)
m2_Q3 = m2.quantile(.75)
m2_IIQ = m2_Q3 - m2_Q1
m2_limite_inferior = m2_Q1 - 1.5 * m2_IIQ
m2_limite_superior = m2_Q3 + 1.5 * m2_IIQ
print(m2_Q1)
print(m2_Q3)
print(m2_IIQ)
print(m2_limite_inferior)
print(m2_limite_superior)
```

```python
data.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = data.groupby('Tipo')['Valor']
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
Q1
```

```python
data_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    e_tipo = data['Tipo'] == tipo
    e_dentro_limite = (data['Valor'] >= limite_inferior[tipo]) & (data['Valor'] <= limite_superior[tipo])
    selecao = e_tipo & e_dentro_limite
    dados_selecao = data[selecao]
    data_new = pd.concat([data_new,dados_selecao])
```

```python
data_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
dados_new.to_csv('Exportações dos Notebooks/aluguel_residencial_sem_outliers.csv', sep= ';', index = False)
```
